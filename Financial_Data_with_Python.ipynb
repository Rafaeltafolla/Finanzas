# Advance Financial data Gathering

```python
#Major Indices
major_indices = pd.read_html('https://finance.yahoo.com/world-indices')[0]
```

```python
major_indices.head()
```

```python
import yfinance as yf

def get_major_indices():
    # List of symbols of the main financial indexes
    major_indices = ['^DJI', '^IXIC', '^GSPC', '^RUT', '^VIX', '^FTSE', '^GDAXI', '^FCHI', '^N225', '^HSI', '^BSESN', '^BVSP', '^MXX']

    # Download YH data
    data = yf.download(major_indices, start='2022-01-01', end='2023-07-26', progress=False)

    return data['Adj Close']

if __name__ == "__main__":
    indices_data = get_major_indices()
    print(indices_data.tail())  # Print the lastest index data
```

```python
#List of the World's major indices
ticker_list = major_indices['Symbol'].str.replace("^","").str.lower().to_list()
```

```python
tickerData = yf.Tickers(ticker_list)
```

```python
len(ticker_list)
```

```python
df = yf.download(ticker_list, period="id", start="2020-01-01", end="2023-07-27")
```

```python
#See the DataFrame
df
```

```python
df.columns
```

```python
df.head()
```

```python
adj_close = df.dropna(thresh=10, axis=1)['Adj Close']
```

```python
adj_close.head()
```

```python
adj_close.describe().T
```

```python
adj_close.plot(figsize=(12,12), subplots=True);
```

# Will drop XAX due to lack of viable data

```python
adj_close = adj_close.drop("XAX", axis=1)
```

```python
adj_close.plot(figsize=(12,20), subplots=True);
```

```python
dji = adj_close['DJI']
```

```python
dji
```

```python
dji.shift(1)
```

```python
dji_per_change = dji / (dji.shift(1) - 1)
```

```python
# Daily percentage change
dji_per_change.plot(figsize=(12,6))
```

```python
# Calculating Yields
# Rate of return
dji_log_returns_shift = np.log(dji / dji.shift(1))
```

```python
dji_log_returns_shift.plot(figsize=(12,6))
```

```python
dji.hist(bins=50,figsize=(12,12));
```
